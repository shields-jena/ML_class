Exercise 2.5: Generating Random Walks

- With N steps in d dimensions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RandWalk (N, d):
  steps = np.random.uniform(-0.5, 0.5, [N,d])
  walks = np.cumsum(steps, axis = 0)
  return walks
```

```python
for _ in range(10): 
    plt.plot(RandWalk(10,1));
plt.title('N=10 steps')
plt.xlabel('step number')
plt.ylabel('walk length')
```

```python
for _ in range(10):
    plt.plot(RandWalk(100,1));
plt.title('N=100 steps')
plt.xlabel('step number')
plt.ylabel('walk length')
```

```python
for _ in range(10):
    plt.plot(RandWalk(1000,1));
plt.title('N = 1,000 steps')
plt.xlabel('step number')
plt.ylabel('walk length')
```

```python
for _ in range(10):
    plt.plot(RandWalk(10000,1));
plt.title('N = 10,000 steps')
plt.xlabel('step number')
plt.ylabel('walk length')
```

```python
for _ in range(10):
    x, y = RandWalk(10,2).transpose()
    plt.plot(x,y);
plt.title('N=10 steps')
plt.xlabel('x axis')
plt.ylabel('y axis')
axes=plt.gca()
axes.set_aspect('equal')
plt.show()
```

```python

for _ in range(10):
    x, y = RandWalk(1000,2).transpose()
    plt.plot(x,y);
plt.title('N=1,000 steps')
plt.xlabel('x axis')
plt.ylabel('y axis')
axes=plt.gca()
axes.set_aspect('equal')
plt.show()
```

```python

plt.figure()
for _ in range(10):
    x, y = RandWalk(100000,2).transpose()
    plt.plot(x,y);
plt.title('N=100,000 steps')
plt.xlabel('x axis')
plt.ylabel('y axis')
axes=plt.gca()
axes.set_aspect('equal')
plt.show()
```

Increasing the step size by 100, roughly increases the net distance by 10 as shown by these xy plots of a random walk. This is seen in how the axes scale by about 10 each time. This is also seen in the 1D random walk, with the axes roughly scaling with $\sqrt{N}$.

b) Find the endpoints of W random walks of N steps and d dimensions:

```python
def Endpoints(W, N, d):
    steps = np.random.uniform(-0.5, 0.5, [W,N,d])
    return np.sum(steps, axis=1)
```

```python
x, y = Endpoints(10000, 10, 2).transpose()
plt.plot(x,y,'.')
x1, y1 = Endpoints(10000, 1, 2).transpose()
plt.plot(x1, y1, '.')
axes=plt.gca()
axes.set_aspect('equal')
plt.title('N=1 and N=10 random walk endpoints')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

This is an example of emergent symmetry as shown by the trends in the distribution that emerge with high W value. Despite a single step having a square distribution, a circular symmetry emerges with greater step numbers.

b) lag function



```python
def P(lag, data):
   """
   Function which returns a list of percentage changes after a number
   "lag" of trading days. 
   P(1) gives the daily percentage changes, P(5) the weekly changes, P[252] the yearly changes, etc.
   To find the ratio of values of an array v of length N after a shift of t, one could use
       v[t:]/v[:-t]
   """
   ratios = data[lag:]/data[:-lag]
   P = 100*(ratios-1)
   return P
```

```python
plt.hist(P(1,SP), bins = 50, density = True)
plt.title('Percentage changes after 1 trading day')
plt.figure()
plt.hist(P(5,SP), bins = 50, density = True)
plt.title('Percentage changes after 1 trading week')
plt.figure()
plt.hist(P(252,SP), bins = 50, density = True)
plt.title('Percentage changes after 1 trading year')
```

The only lag that leads to mean percentage growth greater than zero is the year long lag time. This means investments should be on the year long scale not on a weekly or daily scale. The yearly histogram looks more complex than the daily or weekly one probabably because the data only covers 21 years but we are plotting around 5000 points of yearly percent changes.

c) A Gaussian follows the form of $G  \frac{1}{\sqrt{2}} exp(-x^2/2^2) $.

$log(G) = -log(\frac{1}{\sqrt{2}})  (x^2/2^2)$.

This will be a negative parabola with the log term/$2^2$ being the amplitude. 

```python
lag = 5
plt.hist(P(lag, SP), bins = 50, log=True, density = True)
x = np.arange(-8.,8.5,0.1)
sig = P(lag, SP).std()
x0 = P(lag, SP).mean()
gauss = (1./(np.sqrt(2*np.pi)*sig))*np.exp(-(x-x0)**2/(2*sig**2))
plt.plot(x,gauss,'r');
plt.title('Log of weekly percent change')
```

Compared to the gaussian plot, the log of weekly percent change has fatter tails than expected. There are larger percent changes than expected, especially negative ones.

c) Chemotaxis modeled with biased random walk.


```python
V = 20 
T = 1
T_plus = 1.5
tau = 0.1

N = 1000
time = np.zeros(2*N)
distance = np.zeros(2*N)
signs = np.sign(np.random.uniform(-1, 1, N))

for i in range(N):
  if signs[i] > 0:
    time[2*i] = T_plus
    distance[2*i] = T_plus*V
    
  else:
    time[2*i] = T
    distance[2*i] = -T*V
  time[2*i +1] = tau
  distance[2*i +1] = 0

time_cum = np.cumsum(time)
distance_cum = np.cumsum(distance)

plt.plot(time_cum, distance_cum)
plt.title('Biased random walk')
plt.xlabel('time [s]')
plt.ylabel('Distance [um]')
```

With a biased random walk, the system tends toward the direction which has more favorable conditions. In this case, this is the direction where the step time is longer when it is in the positive direction. In this particular random walk, the average dx/dt is around 4000/1000 ~ 4m/s. This is approximately equal to (T_plus - T)/(T_plus + T) *V.
